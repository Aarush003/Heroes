```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heroes.csv')
print(df)
```

```python
# count how many character by gender
df = df['Gender'].value_counts().reset_index()
print(df)
```

```python
# Drop the entry where Gender is '-'
df = df.loc[df['Gender'] != '-'].reset_index()
print(df)
```

```python
# Reload the original dataframe to ensure 'Alignment' column exists
df = pd.read_csv("heroes.csv")

# count how many heroes are in each gender and alignment category
df = df[(df['Gender'] != '-') & (df['Alignment'] != '-')].reset_index(drop=True)
ndf = df.groupby(['Gender', 'Alignment']).size().reset_index(name='count')
print(ndf)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace with your CSV filename)
df = pd.read_csv("heroes.csv")
 #Visualizations

# Characters by Publisher
df["Publisher"].value_counts().head(10).plot(kind="bar", figsize=(8,4), title="Top 10 Publishers by Number of Characters")
plt.show()

# Alignment distribution
df["Alignment"].value_counts().plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), title="Alignment Distribution")
plt.ylabel("")
plt.show()

# Find the correlation between Height and Weight
df[["Height", "Weight"]].corr()
plt.scatter(df['Height'], df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()

```
